```python
# Define the index name
index_name = "quickstart2"

# Create the vector index
pc.create_index(
    name=index_name,
    dimension=384,  # Replace with your embedding model's output dimension
    metric="cosine",  # Use cosine similarity for vector search
    spec=ServerlessSpec(
        cloud="aws",  # Choose the cloud provider
        region="us-east-1"  # Choose the region
    )
)

# Connect to the index
index = pc.Index(index_name)
print(f"Index '{index_name}' connected successfully!")
```

```python
# Load the embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Sample data
texts = [
    "What is vector search?",
    "Explain Pinecone functionality.",
    "How does Pinecone work for AI applications?",
    "What are embeddings in machine learning?"
]

# Generate embeddings for the sample data
embeddings = model.encode(texts)

# Store embeddings in Pinecone
for i, vector in enumerate(embeddings):
    index.upsert(vectors=[(f"id-{i}", vector)])
print("Data successfully stored in Pinecone!")
```

```python
# Input query
query = "What is Pinecone used for?"

# Generate embedding for the query
query_embedding = model.encode([query])[0].tolist()  # Convert NumPy array to list

# Perform a similarity search
results = index.query(vector=query_embedding, top_k=3)

# Display results
print("Top matches:")
for match in results["matches"]:
    print(f"ID: {match['id']}, Score: {match['score']}")
```
